```python

import pandas as pd
df = pd.read_csv('labels.csv') 
```

#### AUDIO

```python
import pyttsx3

# Initialize the pyttsx3 engine
engine = pyttsx3.init()

# Set the voice
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[0].id)  # You can choose a different voice by changing the index

# Set the text to be spoken
text = "Hello, User! Welcome to traffic sign detection system."

# Convert text to speech
engine.say(text)

# Run the speech engine
engine.runAndWait()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('my_model.h5')

# Load the video stream
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the video stream
    ret, frame = cap.read()
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Preprocess the image
    img = cv2.resize(img, (100,100))
    frame_ = img 
    img = np.expand_dims(img, axis=0)
    
    # Use the model to predict the class
    pred = model.predict(img)
    class_idx = np.argmax(pred)
    

    if class_idx in [25,26,17,16,10,54,55,35,36,28]:

        class_label = df.iloc[class_idx,1]

        engine.say(class_label)
        engine.runAndWait()
    
        # Draw the class label on the image
        cv2.putText(frame, class_label, (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Show the image with the class label
    cv2.imshow('Vehicle Type Detection', frame)
    
    # Wait for a key press
    if cv2.waitKey(1) == ord('q'):
        break

# Release the video stream and close the window
cap.release()
cv2.destroyAllWindows()
```
